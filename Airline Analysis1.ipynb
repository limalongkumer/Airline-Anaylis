```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math

## Read in Data
flight = pd.read_csv("flight.csv")
print(flight.head())

## Task 1
print(flight['coach_price'].mean())
print(flight['coach_price'].median())
sns.histplot(flight['coach_price'])
plt.show()
plt.clf()
```

```python
print(np.mean(flight.coach_price[flight.hours == 8]))
sns.histplot(flight.coach_price[flight.hours == 8])
plt.show()
plt.clf()
```

```python
sns.histplot(flight.delay[flight.delay <= 50])
plt.show()
plt.clf()
```

```python
perc = 0.1
flight_sub = flight.sample(n = int(flight.shape[0]*perc))
sns.lmplot(x='coach_price', y='firstclass_price', data = flight_sub, line_kws={'color': 'black'}, lowess=True)
plt.show()
plt.clf()
```

```python
sns.histplot(flight, x = 'coach_price', hue = flight.inflight_meal)
plt.show()
plt.clf()

sns.histplot(flight, x = 'coach_price', hue = flight.inflight_entertainment)
plt.show()
plt.clf()
sns.histplot(flight, x = 'coach_price', hue = flight.inflight_wifi)
plt.show()
plt.clf()
```

```python
sns.lmplot(x="hours", y="passengers", data=flight, x_jitter=0.25, scatter_kws={'s':3, 'alpha':0.2}, fit_reg=False)
plt.show()
plt.clf()
```

```python
sns.lmplot(x="coach_price", y="firstclass_price", hue = 'weekend', data=flight, fit_reg=False)
plt.show()
plt.clf()
```

```python
sns.boxplot(x = "day_of_week", y = "coach_price", hue = "redeye", palette = 'pastel', data = flight)
plt.show()
plt.clf()
```
